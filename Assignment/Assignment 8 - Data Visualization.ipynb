### Assignment 8 - Data visualization

#### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X_reduced = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target
fig = plt.figure(1,figsize = (8,6))
ax = Axes3D(fig, elev=-150,azim=110)
ax.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2],c=Y,
          cmap=plt.cm.rainbow_r)
ax.set_title("First 3 PCA Directions")
ax.set_xlabel("1st PCA Direction")
ax.set_ylabel("2nd PCA Direction")
ax.set_zlabel("3rd PCA Direction")
```
